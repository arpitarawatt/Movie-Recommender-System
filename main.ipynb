```python
#it is a module
import ast
```

```python
#creating helper function
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
# for top 3
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else: 
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies
```

```python
movies.iloc[0].crew
```

```python
#we'll create a function so that we can only fetch director's name 

def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
    
```

```python
movies.crew=movies['crew'].apply(fetch_director)
```

```python
#overview is in the form of string, we'll convert it into a list so that we can concatenate it with other columns 
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['production_companies']=movies['production_companies'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']+movies['production_companies']
```

```python
selected_df['tags']=selected_df['tags'].apply(lambda x:" ".join(x))
```

```python
selected_df['tags']=selected_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
selected_df['tags']=selected_df['tags'].apply(stem)
```

```python
#VECTORIZATION: Convery the text into vector
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(selected_df['tags']).toarray().shape
```

```python
vectors=cv.fit_transform(selected_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
def recommend(movie):
    movie_index=selected_df[selected_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(selected_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
